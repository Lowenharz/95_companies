```python
# Import necessary libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pprint import pprint
import plotly.graph_objects as go
```

```python
percent_change_monthly_2 =  pd.read_csv('./NASDQ_or_NYSE_percent_change_Monthly_1999_base.csv')
```

```python
f = open('./stock_ticker.txt','r')
file_list = f.readlines()
NYSE_or_NASDAQ = ['^GSPC']
for line in file_list: 
    splited_line =line.split(':') 
    if splited_line[0] in ['NYSE', 'NASDAQ']:
        NYSE_or_NASDAQ.append(splited_line[1].strip())

NYSE_or_NASDAQ.remove('LNKD')
```

```python
percent_change_monthly_2 =  pd.read_csv('./NASDQ_or_NYSE_percent_change_Monthly_1999_base.csv',index_col = 0)
# pprint(percent_change_monthly_2_all)
percent_change_monthly_2_sin_SNP= percent_change_monthly_2.drop('^GSPC',axis =1)
# pprint(percent_change_monthly_2_sin_SNP)
percent_change_monthly_2_sin_SNP = percent_change_monthly_2_sin_SNP -1 
# mean = percent_change_monthly_2_sin_SNP.mean(axis = 0, skipna = True)
percent_change_monthly_2_sin_SNP['mean'] = percent_change_monthly_2_sin_SNP.mean(axis=1)
# pprint(percent_change_monthly_2_sin_SNP)
result = percent_change_monthly_2['^GSPC'] -1
# pprint(percent_change_monthly_2_sin_SNP['mean'])
percent_change_monthly_aggregate = percent_change_monthly_2_sin_SNP['mean']
result = pd.concat([result, percent_change_monthly_aggregate], axis=1, sort=False)
result.to_csv('./NASDQ_or_NYSE_aggregate_percent.csv')
pprint(result)
```

```python
result =  pd.read_csv('./NASDQ_or_NYSE_aggregate_percent.csv')
result.columns = ['Date', 'SNP 500', 'Aggregated_Ethical_Mean']
pprint(result)
```

```python
fig = go.Figure()

stocks = result.columns[1:]
# print(stocks)
for stock in stocks: 
    fig.add_trace(go.Scatter(x=result['Date'], y=result[stock],
                            mode='lines',
                            name=stock))

fig.show()
```
