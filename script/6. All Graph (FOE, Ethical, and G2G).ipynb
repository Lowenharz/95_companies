```python
# Import necessary libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pprint import pprint
import plotly.graph_objects as go
```

```python
G2G =  pd.read_csv('./G2G_aggregate_percent.csv',index_col = 0)
# G2G.columns = ['Date', 'SNP 500', 'Aggregated_G2G_Mean']
# pprint(G2G)
ETHICAL =  pd.read_csv('./NASDQ_or_NYSE_aggregate_percent.csv',index_col = 0)
# ETHICAL.columns = ['Date', 'SNP 500', 'Aggregated_Ethical_Mean']
# pprint(ETHICAL)
FOE =  pd.read_csv('./FOE_aggregate_percent.csv',index_col = 0)
# FOE.columns = ['Date', 'SNP 500', 'Aggregated_FOE_Mean']
# pprint(FOE)
```

```python
result = pd.concat([G2G, ETHICAL['mean'],FOE['mean']], axis=1, sort=False)
result.columns = ['SNP 500', 'Aggregated_G2G_Mean', 'Aggregated_ETHICAL_Mean','Aggregated_FOE_Mean']
result.to_csv('./All_aggregate_percent.csv')
result =  pd.read_csv('./All_aggregate_percent.csv')
# pprint(result)
```

```python
# pprint(result)
fig = go.Figure()

stocks = result.columns[1:]
# print(stocks)
for stock in stocks: 
    fig.add_trace(go.Scatter(x=result['Date'], y=result[stock],
                            mode='lines',
                            name=stock))

fig.show()
```
